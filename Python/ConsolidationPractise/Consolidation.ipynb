```python
account = pd.read_excel('Datafiles/Account Report - Mon Jul 04 2022.xlsx',sheet_name = 2)
asset   = pd.read_excel('Datafiles/Asset Report - Mon Jul 04 2022.xlsx')
emp     = pd.read_excel('Datafiles/Employee Report - Mon Jul 04 2022.xlsx',sheet_name = 4)
emp.head()
```

```python
# identift the columns that we need
account = account[["Firm ID","Account report date","Account report amount"]]
emp = emp[["Firm ID","Date firm established","Employee payment amount"]]
#emp.columns - get the column names

account = account.rename(columns = {
    "Firm ID":"firm_id",
    "Account report date":"date",
    "Account report amount":"amount"
})

emp = emp.rename(columns = {
    "Firm ID":"firm_id",
    "Date firm established":"date",
    "Employee payment amount":"amount"
})

```

```python
print(account.shape,emp.shape)
```

```python
# create a new column "file"
account["file"] = "account"
emp["file"] = "employee"

Final = pd.concat((account,emp))
Final.shape
```
